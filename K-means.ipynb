```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# Fig 3c
# 1. Load the data
data = pd.read_csv("delta_all.csv", sep='\t').values.flatten().tolist()
data = np.array([x for x in data if x!=0])
data_reshaped = data.reshape(-1, 1)

# 2. Perform K-means with N=6
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(data_reshaped)

# 3. Create DataFrame for analysis
df = pd.DataFrame({'value': data, 'cluster': kmeans.labels_})

# 4. Calculate statistics per cluster
stats = df.groupby('cluster')['value'].agg(['count', 'mean', 'std', 'min', 'max'])

# Add centroid information
stats['centroid'] = [kmeans.cluster_centers_[i][0] for i in stats.index]

# Sort by centroid value (low to high) for easier reading
stats = stats.sort_values('centroid')

# Reorder columns
stats = stats[['centroid', 'std', 'count', 'mean', 'min', 'max']]

print(stats)
```
